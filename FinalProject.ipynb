```python
import requests
import pandas as pd

# Replace 'your_api_key' with your actual Twelve Data API key
api_key = '3cc2c2885bea431abf95d6f776acc139'  # Note: API keys should be kept private and not shared in public code
symbol = 'AAPL'

# Make the API request for the income statement with quarterly data
url = f"https://api.twelvedata.com/income_statement?symbol={symbol}&apikey={api_key}&period=quarterly"
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Convert the JSON response to a dictionary
    data = response.json()
    
    # Extract the income statement data
    income_statement_data = data.get('income_statement', [])
    
    # Flatten the data, including nested dictionaries for operating_expense and non_operating_interest
    flattened_data = []
    for item in income_statement_data:
        # Flatten nested dictionaries and combine with the rest of the item
        flattened_item = {**item, **item.get('operating_expense', {}), **item.get('non_operating_interest', {})}
        # Remove the original nested dictionaries
        flattened_item.pop('operating_expense', None)
        flattened_item.pop('non_operating_interest', None)
        flattened_data.append(flattened_item)
    
    # Convert the flattened data to a pandas dataframe
    df_income_statement = pd.DataFrame(flattened_data)
    
    # Optionally, sort the DataFrame by 'fiscal_date' if chronological order is desired
    df_income_statement.sort_values(by='fiscal_date', inplace=True)
    
    # Reset index after sorting
    df_income_statement.reset_index(drop=True, inplace=True)
    
    # Display the dataframe
    print(df_income_statement.head())  # Display the first few rows to check
else:
    print("Failed to retrieve data. Status code:", response.status_code)
df_income_statement.columns
```

```python
import requests
import pandas as pd

# Replace 'your_api_key' with your actual Twelve Data API key
api_key = '3cc2c2885bea431abf95d6f776acc139'
symbol = 'AAPL'
period = 'quarterly'  # Ensure the data is fetched by quarters

# Construct the API URL
url = f'https://api.twelvedata.com/cash_flow?symbol={symbol}&apikey={api_key}&period={period}'

# Make the API request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Convert the JSON response to a pandas DataFrame
    data = response.json()
    
    # Extract 'cash_flow' data
    cash_flow_data = data['cash_flow']
    
    # Normalize the data and create a DataFrame
    df_cash_flow = pd.json_normalize(cash_flow_data)
    
    # Add the 'meta' information as new columns if needed
    for key in data['meta']:
        df_cash_flow[key] = data['meta'][key]
    
    # Sort the DataFrame by 'fiscal_date' to ensure it's in chronological order
    df_cash_flow.sort_values(by='fiscal_date', inplace=True)
    
    # Reset index after sorting
    df_cash_flow.reset_index(drop=True, inplace=True)
    
    print(df_cash_flow.head())  # Display the first few rows of the DataFrame
else:
    print(f"Failed to fetch data: {response.status_code}")
df_cash_flow.columns
```
