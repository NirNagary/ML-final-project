```python
import requests
import pandas as pd
import time
import os



csv_file_name = 'income_statements.csv'


write_header = not os.path.exists(csv_file_name)

# Iterate over each ticker symbol
for symbol in tickers:
    print(f"Processing {symbol}...")
    url = f"https://api.twelvedata.com/income_statement?symbol={symbol}&apikey={api_key}&period=annual"
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200 and response.json().get('status') != 'error':
        data = response.json()
        
        # Extract and flatten the income statement data
        income_statement_data = data.get('income_statement', [])
        flattened_data = []
        for item in income_statement_data:
            flattened_item = {**item, **item.get('operating_expense', {}), **item.get('non_operating_interest', {})}
            flattened_item.pop('operating_expense', None)
            flattened_item.pop('non_operating_interest', None)
            flattened_data.append(flattened_item)
        
        # Convert the flattened data to a DataFrame
        df_income_statement = pd.DataFrame(flattened_data)
        df_income_statement['ticker'] = symbol
        # Append DataFrame to the CSV file
        df_income_statement.to_csv(csv_file_name, mode='a', index=False, header=write_header)
        
        # Ensure the header is not written again in subsequent writes
        write_header = False
        
        print(f"Data for {symbol} added to {csv_file_name}.")
    else:
        print(f"No data for {symbol} or failed to retrieve data. Status code: {response.status_code}")
    
    # Wait for 12 seconds before processing the next ticker
    time.sleep(12)

print("All requested data processed.")
```

```python
csv_file_name = 'cash_flow_data.csv'

# Check if the CSV file already exists to decide on writing headers or not
write_header = not os.path.exists(csv_file_name)

# Iterate over each ticker symbol
for symbol in tickers:
    print(f"Processing {symbol}...")
    url = f'https://api.twelvedata.com/cash_flow?symbol={symbol}&apikey={api_key}&period=annual'
    
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200 and 'cash_flow' in response.json():
        data = response.json()
        
        # Check if cash_flow data is not empty
        if 'cash_flow' in data and data['cash_flow']:
            # Normalize the data and create a DataFrame
            df_cash_flow = pd.json_normalize(data['cash_flow'])
            
            # Add the 'meta' information as new columns, if needed
            if 'meta' in data:
                for key in data['meta']:
                    df_cash_flow[key] = data['meta'][key]
            
            # Check if 'fiscal_date' column exists before sorting
            if 'fiscal_date' in df_cash_flow.columns:
                df_cash_flow.sort_values(by='fiscal_date', inplace=True)
                df_cash_flow.reset_index(drop=True, inplace=True)
            else:
                print(f"'fiscal_date' column not found in data for {symbol}.")
            
            # Append DataFrame to the CSV file
            df_cash_flow.to_csv(csv_file_name, mode='a', index=False, header=write_header)
            
            # Ensure the header is not written again in subsequent writes
            write_header = False
            
            print(f"Data for {symbol} added to {csv_file_name}.")
        else:
            print(f"No cash flow data available for {symbol}.")
    else:
        print(f"Failed to fetch data for {symbol}. Status code: {response.status_code}")
    
    # Wait for 12 seconds before processing the next ticker
    time.sleep(12)

print("All requested data processed.")
```

```python
csv_file_name = 'company_profiles.csv'

# Check if the CSV file already exists to decide on writing headers or not
write_header = not os.path.exists(csv_file_name)

# Iterate over each ticker symbol
for symbol in tickers:
    print(f"Processing {symbol}...")
    url = f"https://api.twelvedata.com/profile?symbol={symbol}&apikey={api_key}"
    
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        
        # Handle cases where the API response might not be what is expected
        if isinstance(data, dict) and not data.get('code'):  # Check if the response is a dictionary and does not contain an error code
            # Convert the dictionary to a DataFrame
            df_profile = pd.DataFrame([data])
            
            # Append DataFrame to the CSV file
            df_profile.to_csv(csv_file_name, mode='a', index=False, header=write_header)
            
            # Ensure the header is not written again in subsequent writes
            write_header = False
            
            print(f"Profile data for {symbol} added to {csv_file_name}.")
        else:
            print(f"No profile data available for {symbol} or error in response.")
    else:
        print(f"Failed to fetch data for {symbol}. Status code: {response.status_code}")
    
    # Wait for 12 seconds before processing the next ticker
    time.sleep(1.5)

print("All requested profile data processed.")
```

```python
csv_file_name = 'stock_splits.csv'

# Check if the CSV file already exists to decide on writing headers or not
write_header = not os.path.exists(csv_file_name)

# Iterate over each ticker symbol
for symbol in tickers:
    print(f"Processing {symbol}...")
    url = f"https://api.twelvedata.com/splits?symbol={symbol}&apikey={api_key}&range=full"
    
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        
        # Extract the splits data, handling cases where 'splits' key may not exist or be empty
        splits_data = data.get('splits', [])
        
        if splits_data:  # Check if there is any splits data
            # Convert the splits data to a DataFrame
            df_splits = pd.DataFrame(splits_data)
            
            # Add a column for the ticker symbol to identify the data
            df_splits['symbol'] = symbol
            
            # Append DataFrame to the CSV file
            df_splits.to_csv(csv_file_name, mode='a', index=False, header=write_header)
            
            # Ensure the header is not written again in subsequent writes
            write_header = False
            
            print(f"Splits data for {symbol} added to {csv_file_name}.")
        else:
            print(f"No splits data available for {symbol}.")
    else:
        print(f"Failed to fetch data for {symbol}. Status code: {response.status_code}")
    
    # Wait for 12 seconds before processing the next ticker
    time.sleep(3)

print("All requested splits data processed.")
```
